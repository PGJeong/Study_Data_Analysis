```python
import pandas as pd
import openpyxl

# 데이터 프레임 만들기 - p.81
# 딕셔너리 자료형을 사용함
df = pd.DataFrame({'Name' : ['JHKim', 'YJLee', 'DHPark', 'MJKim'],
                   'English' : [90, 80, 60, 70],
                   'math' : [50, 60, 100, 20]})
df
```

```python
# 특정 변수의 값 추출하기 - p.83
df['English']
```

```python
# 변수의 값으로 합계 구하기 - p.83
# 내장함수 sum()을 사용함 (<-> 패키지 함수)
sum(df['English'])
```

```python
# 변수의 값으로 평균 구하기 - p.83
# 내장함수 len()을 사용함 (해당 객체의 원소 개수를 반환, 데이터 프레임의 경우 행의 개수)
sum(df['English']) / len(df)
```

```python
# 엑셀 파일을 데이터 프레임으로 변환 - p.86
# 파이썬에서 경로명을 쓸 때는 슬래시(/)를 사용해야 한다
# openpyxl은 파이썬에서 엑셀을 다루기 위해 필요한 패키지
df_exam = pd.read_excel('Data/excel_exam.xlsx')
df_exam
```

```python
# 영어 점수의 전체 평균 구하기 - p.87
sum(df_exam['english']) / len(df_exam)
```
