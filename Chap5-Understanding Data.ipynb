```python
# 실습 - MPG(Mile per Gallon) 데이터 파악하기 - p.105
mpg = pd.read_csv('Data/mpg.csv')
mpg.info()
```

```python
# mpg를 구성하는 11개의 변수 중 숫자로 된 변수 5개의 요약 통계량만 출력됨
mpg.describe()
```

```python
# 모든 변수에 대한 통계량을 출력하려면 include = 'all'을 입력
# unique : 고유값 빈도(중복을 제거한 범주의 개수)
# top : 최빈값(개수가 가장 많은 값)
# freq : 최빈값 빈도(최빈값 개수)
mpg.describe(include = 'all')
```

```python
# 변수의 자료구조 파악 - p.110
type(mpg)
```

```python
# 데이터 프레임의 복사본 만들기 - p.114
# 변수명을 변경하는 등 데이터를 변경할 때 "원본을 보존하기 위해" 복사본 생성(권장사항)
mpg_copy = mpg.copy()
mpg_copy
```

```python
# 변수명 변경하기 - p.114
# 딕셔너리 자료 구조 사용 - df.rename(columns = {'기존 변수명' : '새 변수명'})
mpg_copy = mpg_copy.rename(columns = {'cty' : 'city', 'hwy' : 'highway'})
mpg_copy
```

```python
# 파생변수 만들기 - p.116
# 파생변수 : 변수를 조합하거나 함수를 이용해 만든 새 변수
mpg_copy['total'] = (mpg_copy['city'] + mpg_copy['highway']) / 2
mpg_copy
```

```python
# 통합 연비 변수의 평균 - sum(), len() 내장함수 사용
sum(mpg_copy['total']) / len(mpg)
```

```python
# 통합 연비 변수의 평균 - df.mean() 메서드 사용
mpg_copy['total'].mean()
```

```python
mpg_copy['total'].describe()
```

```python
# 히스토그램으로 데이터 분포 파악하기 - p.119
# df.plot.hist() 메서드
mpg_copy['total'].plot.hist()
```

```python
# 조건문 함수 - p.120
# NumPy 패키지의 where() : 조건문의 true/false 여부에 따라 서로 다른 값 반환
import numpy as np
mpg_copy['test'] = np.where(mpg_copy['total'] >= 20, 'PASS', 'FAIL') # np.where(조건문, true일 때 반환값, false일 때 반환값)
mpg_copy.head()
```

```python
# 빈도표(Frequency Table) - p.121
# 변수의 값이 종류별로 몇 개인지 파악 - df.value_counts()
mpg_copy['test'].value_counts()
```

```python
# 막대그래프 - p.122
# df.plot.bar() 메서드
mpg_copy['test'].value_counts().plot.bar()
```

```python
# 축 이름 회전하기 - p.122
# df.plot.bar() 메서드의 기본값이 축 이름을 90도 회전하게 되어있다
# rot 파라미터에 각도를 입력하여 축 이름을 원하는 각도만큼 회전 가능
mpg_copy['test'].value_counts().plot.bar(rot = 0)
```

```python
# 중첩 조건문 - p.123
mpg_copy['grade'] = np.where(mpg_copy['total'] >= 30, 'A',
                    np.where(mpg_copy['total'] >= 25, 'B',
                    np.where(mpg_copy['total'] >= 20, 'C', 'D')))
# Enter로 줄바꿈 후 열을 맞춰 작성하면 가독성을 높일 수 있다
mpg_copy.head()
```

```python
mpg_copy['grade'].value_counts().plot.bar(rot = 0)
```

```python
# 알파벳 순으로 막대 정렬하기 - p.125
# 빈도표는 이름을 빈도 기준으로 내림차순 정렬한다
# 빈도표를 알파벳 순으로 정렬한 후 그래프를 만들면 알파벳 순으로 정렬 됨
mpg_copy['grade'].value_counts() # 정렬 전
```

```python
mpg_copy['grade'].value_counts().sort_index() # 정렬 후
```

```python
mpg_copy['grade'].value_counts().sort_index().plot.bar(rot = 0)
# 메서드 체이닝(Method Chaining) : 점(.)을 이용해 메서드를 이어서 작성
```

```python
# 여러 조건 중 하나에 해당하면 특정 값을 부여하는 파생변수 - p.128
mpg_copy['size'] = np.where((mpg_copy['category'] == 'compact') |
                            (mpg_copy['category'] == 'subcompact') |
                            (mpg_copy['category'] == '2seater'),
                            'small', 'large')
# vertical bar(|)를 사용하여 'or' 연산가능
# 괄호()를 사용하여 연산우선순위 조정
mpg_copy['size'].value_counts()
```

```python
# 앞의 코드처럼 'or' 조건이 여러번 반복되는 경우 df.isin() 메서드를 사용
# df.isin()에 매개변수로 리스트(조건목록) 삽입
mpg_copy['size'] = np.where(mpg_copy['category'].isin(['compact','subcompact','2seater']),
                           'small', 'large')
mpg_copy['size'].value_counts()
```
